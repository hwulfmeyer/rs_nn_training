```python
import sys, os
import re
import shutil
import pascal_voc_writer
import tarfile
import datetime
from lxml import etree
import tensorflow as tf
from object_detection.utils import dataset_util
```

```python
PATH = "/media/data/LocalizationDataNew/xml"
OUT_PATH = "/media/data/LocalizationDataNew/xml_sorted"

def parseXML(file):
    with tf.gfile.GFile(file, 'r') as fid:
        xml_str = fid.read()
    xml = etree.fromstring(xml_str)
    return dataset_util.recursive_parse_xml_to_dict(xml)

for f in os.listdir(PATH):
    if os.path.isdir(PATH+'/'+f): continue
    #print(f)
    try:
        xml = parseXML(PATH+'/'+f)
        xml_path = xml['annotation']['path']
    except Exception as e:
        print('Can\'t read xml')
    else:
        if 'youbot' in xml_path:
            xml_subdir = xml_path.rsplit('EvalData1',1)[1].rsplit('frame',1)[0]
            date = str(datetime.datetime.fromtimestamp(os.path.getmtime(PATH+'/'+f))).rsplit('.',1)[0]
            comp_path = OUT_PATH+xml_subdir+'/'
            new_name = f.rsplit('_',1)[0] + ' ' + date + '.xml'
            os.makedirs(comp_path, exist_ok=True)
            shutil.copy(PATH+'/'+f, comp_path+new_name)
```

```python
for root, dirs, files in os.walk(OUT_PATH):
    files.sort()
    if len(files) > 0:
        most_recent = files[0]
    for f in files:
        #file = root+'/'+f
        frame,time = f.split(' ',1)
        if most_recent.split(' ',1)[0] == frame:
            most_recent = f
        else:
            print(most_recent.split(' ',1))
            out_root = root.replace('xml_sorted','xml_sorted2')
            os.makedirs(out_root,exist_ok=True)
            shutil.copy(root+'/'+most_recent, out_root+'/'+most_recent.split(' ',1)[0]+'.xml')
            most_recent = f
```
