```python
import os, csv, random
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
```

```python
log_dir = '/media/lukas/My Passport/Dateien/Projekte/2017/ISIC-2017-Data/Results/'
len_avg = 50
marker_size = 5
cap_size = 6
```

```python
def generate_netname(bottleneck_name, setnum, size, dense_num, dropout_rate, optimizer_name, learning_rate):
    net_string = ''
    for i in range(len(dense_num)):
        net_string += str(dense_num[i])+'n'+ \
            str(dropout_rate[i]).replace('.','')+'d'
    return generate_netname_str(bottleneck_name, setnum, size, net_string, optimizer_name, learning_rate)

def generate_netname_str(bottleneck_name, setnum, size, net_string, optimizer_name, learning_rate):
    net_name = bottleneck_name + str(setnum) + '-'
    net_name += str(size) + 'px-'
    net_name += net_string
    net_name += '-'+optimizer_name + str(learning_rate).replace('.','') + 'lr'
    return net_name

def get_avg_valacc(logdir):
    files = []
    for file in os.listdir(logdir):
        files.append(file)
    assert len(files) == 1, 'Multiple files in: ' + logdir + str(files)
    
    val_acc = []
    for e in tf.train.summary_iterator(logdir + files[0]):
        for v in e.summary.value:
            if v.tag == 'val_acc':
                val_acc.append(v.simple_value)
    assert len(val_acc) == 500, 'Log incomplete: ' + logdir
    val_acc = val_acc[-len_avg:]
    assert len(val_acc) == len_avg
    return sum(val_acc)/len(val_acc)

def get_avg_valacc_csv(logdir):
    files = []
    for file in os.listdir(logdir):
        files.append(file)
    #if len(files) != 1: print('Multiple files in: ' + logdir + str(files))
    
    val_acc = []
    with open(logdir + files[0], newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
        first_row = True
        for row in spamreader:
            if first_row==True: 
                first_row = False
                continue
            val_acc.append(float(row[4]))
    assert len(val_acc) == 500, 'Log incomplete: ' + logdir + ' size: ' + str(len(val_acc))
    val_acc = val_acc[-len_avg:]
    assert len(val_acc) == len_avg
    return sum(val_acc)/len(val_acc)
```

```python
## DataSet3 completness check

logname = '171102-VGG16-Dataset3/'
x,y,e = [],[],[]
for nn in range (5,10):
    for size in range(40,201,20):
        for i in range(10):
            for dropout in np.arange(0.1,0.5,0.1):
                try:
                    netname = generate_netname('VGG16-TrainVal3Set',i, size, [2**nn], [dropout], 'sgd', 0.01)
                    print(netname + ': ' + str(get_avg_valacc_csv(log_dir+logname+netname+'/')))
                except FileNotFoundError:
                    print('Missing: ' + netname)
        
```

```python
## 32n04d image size dependency

logname = '171028-VGG16-32n04d/'
x,y,e = [],[],[]
for size in range(40, 201, 20):
    valaccset = []
    for setnum in range (1, 11):
        netname = generate_netname('VGG16-TrainValSet',setnum, size, [32], [0.4], 'sgd', 0.005)
        valaccset.append(get_avg_valacc(log_dir+logname+netname+'/'))
    x.append(size)
    y.append(sum(valaccset)/len(valaccset))
    e.append(np.std(valaccset))
    
plt.errorbar(x,y,e, fmt='--o', markersize=marker_size, capsize=cap_size)
plt.ylabel('Validation accuracy')
plt.xlabel('Image size')
plt.title('Pretrained VGG16 with 32 neurons topnet')
plt.show()
```

```python
## 32n32n32n dropout dependency @160px
log_name = '171028-VGG16-32n32n32n/'
size = 160

x,y,e = [],[],[]
for dropout in range(2, 5):
    valaccset = []
    for setnum in range (1, 11):
        netname = generate_netname_str('VGG16-TrainValSet',setnum, size, '32n32n32n0'+str(dropout)+'d', 'sgd', 0.005)
        valaccset.append(get_avg_valacc(log_dir+log_name+netname+'/'))
    x.append(dropout*0.1)
    y.append(sum(valaccset)/len(valaccset))
    e.append(np.std(valaccset))
```

```python
# Single configuration

logname = '171102-VGG16-Dataset3/'
nn = 32
size = 200
dropout = 0.9
N = 1

val_acc = []
for i in range(10):
    val_acc.append([])
    netname = generate_netname('VGG16-TrainVal3Set',i, size, [nn], [dropout], 'sgd', 0.01)
    logdir = log_dir+logname+netname+'/'
    files = []
    for file in os.listdir(logdir):
        files.append(file)
    assert len(files) == 1, 'Multiple files in: ' + logdir + str(files)
    
    with open(logdir + files[0], newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
        first_row = True
        for row in spamreader:
            if first_row==True: 
                first_row = False
                continue
            val_acc[i].append(float(row[4]))
val_avg = np.mean(val_acc, axis=0)
val_avg = np.convolve(val_avg, np.ones((N,))/N, mode='valid')

plt.plot(val_avg)
plt.ylabel('Validation accuracy')
plt.xlabel('Epoch')
plt.show()

plt.plot(np.swapaxes(val_acc, 0, 1)[:,8])
plt.ylabel('Validation accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower left')
plt.show()
```
